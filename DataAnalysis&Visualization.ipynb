```python
#2.4
# For each state identify the districts with highest population
District_Demographics
highest_population = District_Demographics[["State","District","Population"]]
highest_population = highest_population.loc[highest_population.groupby("State")["Population"].idxmax()]
highest_population
```

```python
#create a column chart
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(12,6))

# generate unique colors
colors = plt.cm.tab20.colors

for i, (state, pop) in enumerate(zip(highest_population["State"], highest_population["Population"])):
    plt.bar(i, pop, color=colors[i % len(colors)], label=state)

plt.ylabel("Population")
plt.title("District with Highest Population in Each State")

# remove bottom labels
plt.xticks(range(len(highest_population)), [])

# top axis for district names
ax = plt.gca()
top_axis = ax.secondary_xaxis('top')
top_axis.set_xticks(range(len(highest_population)))
top_axis.set_xticklabels(highest_population["District"], rotation=82)

plt.legend(title="States", bbox_to_anchor=(1.05,1), loc="upper left")

plt.show()
```

```python
#5.1
#line chart showing t.number of transactions and t.transaction amount over time( years and quarters) for a selected state 
State_TxnSplit
Andaman_ = State_TxnSplit[State_TxnSplit["State"]=="Andaman & Nicobar Islands"]
Andaman_ = Andaman_.groupby(["Year","Quarter"])[["Transactions","Amount (INR)"]].sum()
Andaman_= Andaman_.reset_index()
Andaman_["Period"] = Andaman_["Year"].astype(str)+ " " +"Q" + Andaman_["Quarter"].astype(str)
Andaman_
fig, ax1 = plt.subplots(figsize=(10,6))

ax1.plot(Andaman_["Period"], Andaman_["Transactions"],color="orange",marker="o", label="Transactions")

ax2 = ax1.twinx()
ax2.plot(Andaman_["Period"], Andaman_["Amount (INR)"], marker="^", label="Amount")

ax1.set_xticks(range(len(Andaman_["Period"])))
ax1.set_xticklabels(Andaman_["Period"], rotation=90)

plt.title("Total Transactions & Amount over Years in Andaman")

fig.legend()

plt.show()
```

```python
#5.2
#pie chart showing distributions of different transaction types for a specific state and quarter
State_TxnSplit
trans_types = State_TxnSplit[(State_TxnSplit["State"]=="Uttar Pradesh") & (State_TxnSplit["Year"]==2020)]
trans_types = trans_types[trans_types["Quarter"]==4]
import matplotlib.pyplot as plt
values = trans_types["Transactions"]
labels = trans_types["Transaction Type"]
explode = [0,0,0,0.35,0]
plt.title("Distribution of various Transaction Types",fontsize=12)
plt.pie(values, labels=labels, explode=explode,autopct = "%1.1f%%")
plt.show()
```
